```python
import numpy as np
from sklearn import linear_model

#x = np.array([1,2,2,3,3,4,5,6,6,6,8,10])
#y = np.array([-890, -1411, -1560, -2220, -2091, -2878, -3537, -3268, -3920, -4163, -5471, -5157])

x = np.array([3,1,0,4])
y = np.array([2,2,1,3])

x = x.reshape(-1,1)
y = y.reshape(-1,1)

regr = linear_model.LinearRegression()
regr.fit(x, y)

b0 = regr.intercept_
print('Intercept c:', b0[0])

b1 = regr.coef_
print('Slope m:', b1[0][0])
```

```python
import numpy as np
from sklearn import linear_model

x = np.array([1,2,2,3,3,4,5,6,6,6,8,10])
y = np.array([-890, -1411, -1560, -2220, -2091, -2878, -3537, -3268, -3920, -4163, -5471, -5157])

x = x.reshape(-1,1)
y = y.reshape(-1,1)

regr = linear_model.LinearRegression()
regr.fit(x, y)

b0 = regr.intercept_
print('Intercept c:', b0[0])

b1 = regr.coef_
print('Slope m:', b1[0][0])
                        
regr.predict(6)
```
