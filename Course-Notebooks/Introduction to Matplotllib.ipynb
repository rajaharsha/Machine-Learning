If you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0,1,2,3].

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')  # default is "-" solid line
plt.axis([0, 6, 0, 20])
plt.show()
```

The axis() command in the example above takes a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes.

#### Working with text
The text() command can be used to add text in an arbitrary location, and the xlabel(), ylabel() and title() are used to add text in the indicated locations. Any text element can use math text. You should use raw strings (precede the quotes with an 'r'), and surround the math text with dollar signs ($), as in TeX.

```python
plt.text?
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, _ = plt.hist(x, 50, normed=1)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

#### Annotating text
A common use for text is to annotate some feature of the plot, and the annotate() method provides helper functionality to make annotations easy. In an annotation, there are two points to consider: the location being annotated represented by the argument xy and the location of the text xytext. Both of these arguments are (x,y) tuples.

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

#### Plot Heat Map - Correlation Matrix

```python
import pandas as pd, pandas_datareader.data as data
from datetime import datetime as dt
```

```python
# give a sample start date and stocks as input
start = dt(2016,10,31)
stocks = ['FB', 'AMZN', 'GOOG', 'NFLX', 'GLD']
```

```python
# read stock closing values from Google
stocks_data = data.DataReader(stocks, 'yahoo', start)['Close']
stocks_data.head()
```

```python
stocks_data.info()
```

```python
# calculate percentage change using pandas functions pct_change()
stocks_pct_change = stocks_data.pct_change()
stocks_pct_change.head()
```

```python
# calculate Pearson Correlation using pandas function corr()
corr_stocks = stocks_pct_change.corr()
corr_stocks
```

```python
plt.figure(figsize=(12, 10))
plt.imshow(corr_stocks, cmap='RdYlGn', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr_stocks)), corr_stocks.columns, rotation='vertical')
plt.yticks(range(len(corr_stocks)), corr_stocks.columns);
plt.suptitle('Stock Correlations Heat Map', fontsize=15, fontweight='bold')
plt.show()
```
