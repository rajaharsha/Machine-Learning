```python
from sklearn.datasets import load_iris 
from sklearn import preprocessing
import numpy as np
from numpy import linalg as LA

iris = load_iris() 
 
numSamples, numFeatures = iris.data.shape 
X = iris.data

Xn = preprocessing.scale(X)
```

```python
Xn.shape
```

```python
u,s,v = np.linalg.svd(Xn.T)
```

```python
u # eigen vector
```

```python
s # eigen value
```

```python
def project_data(X, U, k):  
    U_reduced = U[:,:k]
    return np.dot(X, U_reduced)

def recover_data(Z, U, k):  
    U_reduced = U[:,:k]
    return np.dot(Z, U_reduced.T)
```

```python
Z = project_data(Xn, u, 2)  
print ('2D Reduced Data: \n')
print (Z)
```

```python
X_recovered = recover_data(Z, u, 2)  
print ('Reconstructed Data: \n')
print (X_recovered)
```

```python
print ('Approximation Error: \n')
print (LA.norm(X_recovered-Xn))
```
